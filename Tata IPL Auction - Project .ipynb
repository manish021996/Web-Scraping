```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.iplt20.com/auction/2023'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'lxml')
```

```python
data = soup.find('table', {'class':'ih-td-tab auction-tbl'})
```

```python
header = data.find_all('th')
```

```python
title = []
for i in header:
    title.append(i.text)
```

```python
title
```

```python
df = pd.DataFrame(columns = title)
```

```python
rows = data.find_all('tr')
```

```python
rows
```

```python
for i in rows[1:]:
    name = i.find('h2')
    print(name.text)
    
```

```python
for i in rows[1:]: 
    
    data_lst = []   
    name = i.find('h2')
    data_lst.append(name.text)
    
    other_data = i.find_all('td')
    
    for j in other_data[1:]:
        data_lst.append(j.text)
    
    l = len(df)
    df.loc[l] = data_lst
```

### 2nd Approach 

```python
second_df = pd.DataFrame(columns = title)
```

```python
for i in rows[1:]:
    row = i.find_all('td')
    
    data_list = []
    for j in row:
        data_list.append(j.text)
        
    l = len(second_df)
    second_df.loc[l] = data_list
```

```python
def remove_n(text):
    clean_text = text.replace('\n','')
    return clean_text
```

```python
second_df['TEAM'] = second_df['TEAM'].apply(remove_n)
```

## Top Buy Players Table

```python
top_buy = soup.find('table',class_ = 'ih-td-tab auction-tbl', id = "t4")
```

```python
top_buy
```

```python
top_buy_header = top_buy.find_all('th')

header_list = []
for i in top_buy_header:
    header_list.append(i.text)
```

```python
top_buys_df = pd.DataFrame(columns = header_list)
```

```python
top_buy_rows = top_buy.find_all('tr')
```

```python
top_buy_rows
```

```python
for i in top_buy_rows[1:]:
    top_buy_data = []
    name = i.find('h2')
    top_buy_data.append(name.text)
    
    top_buy_other_data = i.find_all('td')
    for j in top_buy_other_data[1:]:
        top_buy_data.append(j.text)
    
    l = len(top_buys_df)
    top_buys_df.loc[l] = top_buy_data
```

### 2nd Approach

```python
top_buys_df_2 = pd.DataFrame(columns = header_list)

for i in top_buy_rows[1:]:
    row = i.find_all('td')
    
    tb_other_data = []
    for j in row:
        tb_other_data.append(j.text)
    
    l = len(top_buys_df_2)
    top_buys_df_2.loc[l] = tb_other_data

top_buys_df_2["TEAM"] = top_buys_df_2["TEAM"].apply(remove_n)
```
